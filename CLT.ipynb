***Central Limit Theorem : Astrostats 2016***

Dadas $N$ variables aleatorias $X_n$ IID con varianza $\sigma^2$ y media $\mu$ finitas, el promedio $S_N = (1/N) \sum_{i=1}^N X_n$ tiende a una normal $N(\mu,\sigma^2/N)$.

```python
from numpy.random import gumbel
from numpy import squeeze
from numpy import pi, exp, sqrt
from pylab import plot, hold, hist
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
def sample(N, sample_size=100000):
    X_n = gumbel(1.5, 3.0, [sample_size, N])
    return X_n[:,:].mean(1)

def show_gaussian_fit(sample, bins):
    mu = sample.mean()
    sigma = sample.std()
    a = sample.size * (bins[1] - bins[0])
    hold(True)
    plot(bins, a/(sqrt(2*pi)*sigma) * exp(-(bins-mu)**2 / (2*sigma**2)), 'r-')
    return
```

```python
# Plot the distribution of X_n.
from pylab import plot, linspace, gca, xlabel, ylabel
bins = linspace(-5, 15, 100);
X_n = sample(1)
hist(X_n, bins);
show_gaussian_fit(X_n, bins);

# Label plot.
xlabel('x');
ylabel('p(x)');
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
```

claramente, la distribucion de X no es Gaussiana. Pero si empezamos a sumar variables, la distribucion comienza a verse Gaussiana!

```python
S_N = sample(3);
hist(S_N, bins);
show_gaussian_fit(S_N, bins);
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
```

```python
S_N = sample(6);
hist(S_N, bins);
show_gaussian_fit(S_N, bins);
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
```

Con 20 samples, la distribucion es esencialmente indistinguible de una Gaussiana

```python
S_N = sample(20);
hist(S_N, bins);
show_gaussian_fit(S_N, bins);
gca().axes.get_yaxis().set_ticks([]); # turn off y ticks
```
